```python
import os
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
```

```python


# Read in a plain text file
with open(os.path.join("data3", "java.txt"), "r") as f:
    text = f.read()
    #print(text)
```

```python
# Split text into words using NLTK
words = word_tokenize(text)
#print(words)
```

```python
# Remove stop words
words = [w for w in words if w not in stopwords.words("english")]
```

```python
# Reduce words to their root form
lemmed = [WordNetLemmatizer().lemmatize(w) for w in words]
```

```python
#For every term in the current document, and every document in the set, compute the tf-idf
tvec = TfidfVectorizer(min_df=.0025, max_df=.2, stop_words='english', ngram_range=(1,2))
tvec_weights = tvec.fit_transform(lemmed)
weights = np.asarray(tvec_weights.mean(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': tvec.get_feature_names(), 'weight': weights})
#Choose the top 20 words by tf-idf score
weights_df.sort_values(by='weight', ascending=False).head(20)
```

```python

weights_df.to_csv('data3/weights_df.csv', index=False)
```
